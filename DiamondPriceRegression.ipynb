```python
clarity_dict = {'IF':7,'VVS1':6,'VVS2':5,'VS1':4,'VS2':3,'SI1':2,'SI2':1,'I1':0}

df['clarity'].replace(clarity_dict,inplace=True)

df.head()
```

```python
df.info()
```

```python
df['cut'].unique()
```

```python
df = pd.get_dummies(df,columns=['cut','color'])
df.head()
```

```python
NUMERIC_FEATURES = ['carat','depth','table','x','y','z']

numeric_df = df[NUMERIC_FEATURES]

numeric_df.describe()
```

```python
df.info()
```

```python
from sklearn.preprocessing import scale

numeric_array = scale(numeric_df)

numeric_array[:3]
```

```python
numeric_df = pd.DataFrame(numeric_array,columns=numeric_df.columns)

numeric_df.describe()
```

```python
df.drop(NUMERIC_FEATURES,axis=1,inplace=True)
```

```python
df = pd.concat([df,numeric_df],axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('price',axis=1)
y = df['price']

X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression()

clf.fit(X_train,y_train)
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
clf.score(X_train,y_train)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test,y_preds)

r2
```

```python
df.head()
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()

lasso.fit(X_train,y_train)

l_preds = lasso.predict(X_test)

print(r2_score(y_test,y_preds))
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge()

ridge.fit(X_train,y_train)

r_preds = ridge.predict(X_test)

print(r2_score(y_test,y_preds))
```

```python
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier()

sgd.fit(X_train,y_train)

s_preds = sgd.predict(X_test)

print(r2_score(y_test,y_preds))
```
